## Limpeza e Transformação

```python
# Função para detectar duplicatas
def duplicatas(dataf, coluna, nd):
    sit = dataf[coluna].duplicated().any()
    
    
    if sit:
        print('{}[{}] possui valores duplicados.      '.format(nd, coluna))
        duplicadas = dataf[coluna].duplicated(keep=False)
        print(sum(duplicadas))
    else:
        print('{}[{}] Não possui valores duplicados.  '.format(nd, coluna))
```

```python
#Buscar valores duplicados
duplicatas(musics, 'index', 'musics')
duplicatas(musics, 'SongID', 'musics')
duplicatas(features, 'index', 'features')
duplicatas(features, 'SongID', 'features')
```

```python
# Análise de total de valores nulos

# Função para verificar nulos em colunas
def totalNull(df, colunas):
    totNulos = []
    for coluna in colunas:
        nulos = df[coluna].isnull()
        totNulos.append(sum(nulos))

    print(totNulos)


colunas = ['SongID', 'Performer', 'Song', 'spotify_genre', 'spotify_track_id', 'spotify_track_preview_url',
           'spotify_track_duration_ms', 'spotify_track_popularity', 'spotify_track_explicit', 'danceability', 'energy',
           'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 
           'time_signature']

totalNull(features,colunas)
```

```python
# Eliminação dos valores nulos, baseado em 'spotify_track_duration_ms'
features = features.dropna(subset='spotify_track_duration_ms')
```

```python
# Visualizar novo features
features.head()
```

```python
# Verificar Resultado da remoção de nulos
totalNull(features, colunas)
```

```python
# Salvar novo features no diretório local
features.to_csv('Dados/Novofeatures.csv', index=False)
```
